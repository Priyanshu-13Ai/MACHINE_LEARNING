```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#ploating a simple function
price=[33000,40000,50000,49000,46000,37000]
year=[2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
#from a pandas dataFrame
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('rohit vs virat ipl comparision')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#color(hex) and line (width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')

plt.title('rohit vs virat ipl comparision')
plt.xlabel('season')
plt.ylabel('runs scored')

```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted')

plt.title('rohit vs virat ipl comparision')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted')

plt.title('rohit vs virat ipl comparision')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted')

plt.title('rohit vs virat ipl comparision')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',label='rohit')

plt.title('rohit vs virat ipl comparision')
plt.xlabel('season')
plt.ylabel('runs scored')
plt.legend(loc='best')
```

# **BAR GRAPH / BAR PLOT**

```python
#Simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.xlabel('colors')
plt.ylabel('children')

plt.bar(colors,children,color='black')
```

```python
#horizontal bar
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.xlabel('colors')
plt.ylabel('children')

plt.barh(colors,children,color='black') #barh se horizontal
```

```python
#color and label
df=pd.read_csv('/content/sample_data/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
plt.show()
```

```python
#groupby ka best example and work
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='black')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red ','blue blue blue blue','green green green green','yellow yellow yellow yellow','pink']
plt.xlabel('colors')
plt.ylabel('children')

plt.bar(colors,children,color='black')

plt.xticks(rotation=90)
plt.show() # har jagha chalega isse or unwanted info hide ho jayegi
```

# **PIE CHART**

Uni/bi analysis               
categorial vs numerical            
use case - to find contribution on a standard scale

```python
# simple data

data = [23,45,100,20,49]
subjects = ['eng','maths','science','hindi','social']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# Dataset
df = pd.read_csv('/content/sample_data/gayle-175.csv')
df
```

```python

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# Percentage and color
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```
